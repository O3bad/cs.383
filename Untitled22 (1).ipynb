```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import glob
import xml.etree.ElementTree as ET
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Flatten, Conv2D, MaxPooling2D, Dense

path = kagglehub.dataset_download("andrewmvd/car-plate-detection")
print("Path to dataset files:", path)
```

```python
IMAGE_SIZE = 224

def resizeannotation(f):
    tree = ET.parse(f)
    root = tree.getroot()

    width = int(root.find('size').find('width').text)
    height = int(root.find('size').find('height').text)

    obj = root.find('object')
    bndbox = obj.find('bndbox')
    xmin = int(bndbox.find('xmin').text)
    ymin = int(bndbox.find('ymin').text)
    xmax = int(bndbox.find('xmax').text)
    ymax = int(bndbox.find('ymax').text)

    new_xmin = (xmin / width) * IMAGE_SIZE
    new_ymin = (ymin / height) * IMAGE_SIZE
    new_xmax = (xmax / width) * IMAGE_SIZE
    new_ymax = (ymax / height) * IMAGE_SIZE

    return [new_xmin, new_ymin, new_xmax, new_ymax]
```

```python
img_dir = os.path.join(path, "images")
data_path = os.path.join(img_dir, '*g')
files = glob.glob(data_path)
files.sort()

X = []
for f1 in files:
    img = cv2.imread(f1)
    if img is not None:
        img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        blurred = cv2.GaussianBlur(gray, (5, 5), 0)

        edges = cv2.Canny(blurred, 100, 200)

        processed_img = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
        X.append(processed_img)

X = np.array(X)

annotations_dir = os.path.join(path, 'annotations')
text_files = [os.path.join(annotations_dir, f) for f in sorted(os.listdir(annotations_dir)) if f.endswith('.xml')]

y = []
for i in text_files:
    y.append(resizeannotation(i))
y = np.array(y)

print(f"تم تجهيز {len(X)} صورة و {len(y)} إحداثيات.")
```

```python
num_samples = 5

plt.figure(figsize=(15, 10))

for i in range(num_samples):
    original_img = cv2.imread(files[i])
    original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
    original_img = cv2.resize(original_img, (IMAGE_SIZE, IMAGE_SIZE))

    plt.subplot(num_samples, 2, 2*i + 1)
    plt.imshow(original_img)
    plt.title(f"Original Image {i+1}")
    plt.axis('off')

    processed_img = (X[i] * 255).astype('uint8')

    plt.subplot(num_samples, 2, 2*i + 2)
    plt.imshow(processed_img)
    plt.title(f"After Canny Edge Detection {i+1}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
X = X.astype('float32') / 255.0
y = y.astype('float32') / 224.0

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
model = Sequential([
    Conv2D(64, (3,3), input_shape=(224,224,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.1),

    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.1),

    Flatten(),
    Dense(64, activation='relu'),
    Dense(4, activation='sigmoid')
])

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

print("Start Training")
train = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, batch_size=32)
```

```python
idx = 0
img_to_test = X_test[idx].copy()
prediction = model.predict(np.expand_dims(img_to_test, axis=0))[0]

p = prediction * 224
t = y_test[idx] * 224

img_display = (img_to_test * 255).astype('uint8')
cv2.rectangle(img_display, (int(p[0]), int(p[1])), (int(p[2]), int(p[3])), (0, 255, 0), 2)
cv2.rectangle(img_display, (int(t[0]), int(t[1])), (int(t[2]), int(t[3])), (255, 0, 0), 2)

plt.imshow(img_display)
plt.title("Green: Predicted | Red: Ground Truth")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(train.history['accuracy'], label='Train Accuracy', color='blue', linewidth=2)
plt.plot(train.history['val_accuracy'], label='Validation Accuracy', color='orange', linewidth=2)
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(train.history['loss'], label='Train Loss', color='red', linewidth=2)
plt.plot(train.history['val_loss'], label='Validation Loss', color='green', linewidth=2)
plt.title('Model Loss (MSE)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
